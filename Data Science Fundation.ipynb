### Description:
This dataset contains information about customer interactions, sales activities, and opportunities from a fictional company's CRM (Customer Relationship Management) system. The dataset is designed to help data scientists and analytics understand the sales process, identify trends and patterns, and build predictive models to improve sales performance.

### Features:

Customer information (demographics, firmographics, etc.)
Sales activities
Opportunity data (deal size, stage, probability, etc.)
Product/service information
Sales team and performance metrics
Time-series data (daily/weekly/monthly sales, etc.)

### Use Cases:
Predicting won/lost opportunities

Forecasting deal size

Identifying key drivers of sales performance

Optimizing sales team performance

Analyzing customer behavior and preferences

This dataset is perfect for data scientists, analysts, and students looking to practice their skills in:

Predictive modeling

Data visualization

Sales analytics

Customer relationship management

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('C:/Users/menez/OneDrive/Desktop/DATA SCIENCE FUNDATIONS/accounts.csv')
df1
```

```python
print("Contas da Base de Dados:")
print(df1.info())
print(df1.head())
```

```python
acc_null_sum = df1.isnull().sum()
print("Soma Nulos em Toda a Base de Dados:")
print(acc_null_sum)
```

```python
df1['subsidiary_of'] = df1['subsidiary_of'].fillna('Desconhecido')
acc_null_sum = df1.isnull().sum()

# Print null sums
print("Null Sums for Accounts Dataset:")
print(acc_null_sum)
```

```python
valores_unicos = df1['subsidiary_of'].unique()
print(f"Valores únicos na coluna: {valores_unicos}")
```

```python
plt.figure(figsize=(5, 3))
sns.countplot(x='subsidiary_of', data=df1)
plt.title('Contagem de subsidiária')
plt.xlabel('Subsidiária')
plt.ylabel('Contar')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.histplot(df1['revenue'], bins=20, kde=True)
plt.title('Distribuição de Receita')
plt.xlabel('Receita (Em Milhões de Dolares)')
plt.ylabel('Frequencia')
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.boxplot(x='sector', y='employees', data=df1)
plt.title('Distribuição de Funcionários por Setor')
plt.xlabel('Setor')
plt.ylabel('Numero de Funcionários')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.histplot(x='year_established', data=df1, bins=20, kde=True)
plt.title('Distribuição por Ano')
plt.xlabel('Ano')
plt.ylabel('Frequencia')
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.boxplot(x='sector', y='revenue', data=df1)
plt.title('Receita distribuida por Setor')
plt.xlabel('Setor')
plt.ylabel('Receita')
plt.xticks(rotation=80)
plt.show()
```

```python
df_pipeline = pd.read_csv('C:/Users/menez/OneDrive/Desktop/DATA SCIENCE FUNDATIONS/sales_pipeline.csv')
df_pipeline
```

```python
df_pipeline.info()
```

```python
pip_null_sum = df_pipeline.isnull().sum()
print("Soma Nulos em Toda a Base de Dados:")
print(pip_null_sum)
```

```python
df_pipeline['account'] = df_pipeline['account'].fillna('Desconhecido')
df_pipeline['engage_date'] = df_pipeline['engage_date'].fillna('Desconhecido')
df_pipeline['close_date'] = df_pipeline['close_date'].fillna('Desconhecido')
df_pipeline['close_value'] = df_pipeline['close_value'].fillna(0)

pip_null_sum = df_pipeline.isnull().sum()

print("\nNull Sums for Sales Pipeline Dataset:")
print(pip_null_sum)
```

```python
plt.figure(figsize=(6, 3))
sns.boxplot(x='product', y='close_value', data=df_pipeline)
plt.title('Distribuição produtos por valor')
plt.xlabel('Produtos')
plt.ylabel('Valores')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualizing Sales Teams Dataset
plt.figure(figsize=(5, 3))
sns.countplot(x='deal_stage', data=df_pipeline)
plt.title('Contagem Etapas do Negocio')
plt.xlabel('Etapas')
plt.ylabel('Contar')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualizing Sales Teams Dataset
plt.figure(figsize=(10, 6))
sns.countplot(x='sales_agent', data=df_pipeline)
plt.title('Contagem Agentes de Vendas')
plt.xlabel('Agentes')
plt.ylabel('Contar')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(x='sales_agent', hue='product', data=df_pipeline)
plt.title('Contagem de Agentes de Produto')
plt.xlabel('Agentes de Vendas')
plt.ylabel('Contagem')
plt.xticks(rotation=90)
plt.legend(title='Produtos')
plt.show()
```

```python
# Visualizing Sales Pipeline Dataset
plt.figure(figsize=(5, 3))
df_pipeline['deal_stage'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribuição das Etapas do Negócio')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(x='close_value', data=df_pipeline, bins=20, kde=True)
plt.title('Distribuição de valores próximos')
plt.xlabel('Valor Fechado')
plt.ylabel('Frequencia')
plt.show()
```
